```python
# !pip install ipywidgets matplotlib
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
# Define a function for interactive line plot
def interactive_line_plot(frequency):
    x = np.linspace(0, 10, 100)
    y = np.sin(frequency * x)
    plt.figure(figsize=(10, 5))
    plt.plot(x, y)
    plt.title(f'Sine Wave with Frequency {frequency}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()

# Import the interact function from ipywidgets
from ipywidgets import interact, widgets

# Create an interactive slider for the frequency
interact(interactive_line_plot, frequency=widgets.FloatSlider(min=0.1, max=5.0, step=0.1, value=1.0, description='Frequency'));
```

```python
# Define a function for interactive heatmap
def interactive_heatmap(scale):
    data = np.random.rand(10, 10) * scale
    plt.figure(figsize=(10, 5))
    plt.imshow(data, cmap='viridis', interpolation='nearest')
    plt.colorbar()
    plt.title(f'Heatmap with Scale {scale}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()

# Create an interactive slider for the scale
interact(interactive_heatmap, scale=widgets.FloatSlider(min=0.5, max=5.0, step=0.5, value=1.0, description='Scale'));
```

```python
# Define a function for interactive polar plot
def interactive_polar_plot(factor):
    theta = np.linspace(0, 2 * np.pi, 100)
    r = np.abs(np.sin(factor * theta))

    plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, projection='polar')
    ax.plot(theta, r)
    plt.title(f'Polar Plot with Factor {factor}')
    plt.show()

# Create an interactive slider for the factor
interact(interactive_polar_plot, factor=widgets.FloatSlider(min=0.5, max=5.0, step=0.5, value=1.0, description='Factor'));
```

```python
# Define a function for interactive box plot
def interactive_box_plot(width):
    data = [np.random.normal(0, std, 100) for std in range(1, 4)]
    plt.figure(figsize=(10, 5))
    plt.boxplot(data, widths=width)
    plt.title(f'Box Plot with Width {width}')
    plt.xlabel('Categories')
    plt.ylabel('Values')
    plt.show()

# Create an interactive slider for the width
interact(interactive_box_plot, width=widgets.FloatSlider(min=0.1, max=1.0, step=0.1, value=0.5, description='Box Width'));
```

```python
# Import additional library for 3D plotting
from mpl_toolkits.mplot3d import Axes3D

# Define a function for interactive 3D plot
def interactive_3d_plot(angle):
    x = np.linspace(-5, 5, 100)
    y = np.linspace(-5, 5, 100)
    X, Y = np.meshgrid(x, y)
    Z = np.sin(np.sqrt(X**2 + Y**2))

    fig = plt.figure(figsize=(10, 5))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis')
    ax.view_init(30, angle)
    plt.title('Interactive 3D Plot')
    plt.show()

# Create an interactive slider for the angle
interact(interactive_3d_plot, angle=widgets.IntSlider(min=0, max=360, step=5, value=0, description='Angle'));
```

```python
# Define a function for interactive pie chart
def interactive_pie_chart(scale):
    labels = ['A', 'B', 'C', 'D']
    sizes = [15*scale, 30*scale, 45*scale, 10*scale]
    plt.figure(figsize=(8, 8))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%')
    plt.title('Interactive Pie Chart')
    plt.show()

# Create an interactive slider for the scale
interact(interactive_pie_chart, scale=widgets.FloatSlider(min=0.5, max=3.0, step=0.1, value=1.0, description='Scale'));
```

```python
# Define a function for interactive bar chart
def interactive_bar_chart(height):
    labels = ['A', 'B', 'C', 'D']
    plt.figure(figsize=(10, 5))
    plt.bar(labels, [height, height*1.5, height*0.75, height*2])
    plt.title('Interactive Bar Chart')
    plt.xlabel('Categories')
    plt.ylabel('Values')
    plt.show()

# Create an interactive slider for the height
interact(interactive_bar_chart, height=widgets.FloatSlider(min=0.5, max=10, step=0.5, value=1.0, description='Height'));
```

```python
# Define a function for interactive histogram
def interactive_histogram(bins):
    data = np.random.randn(1000)
    plt.figure(figsize=(10, 5))
    plt.hist(data, bins=bins)
    plt.title(f'Histogram with {bins} Bins')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()

# Create an interactive slider for the bins
interact(interactive_histogram, bins=widgets.IntSlider(min=5, max=50, step=1, value=10, description='Bins'));
```

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Example 1: Interactive sine wave with frequency control
def plot_sine_wave(freq):
    x = np.linspace(0, 10, 100)
    y = np.sin(freq * x)

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, '-b')
    plt.title(f'Interactive Sine Wave (Frequency: {freq})')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

interact(plot_sine_wave, freq=(1, 10, 0.5))
```

```python
# Example 2: Scatter plot with adjustable points
def plot_scatter(n_points):
    x = np.random.rand(n_points)
    y = np.random.rand(n_points)
    colors = np.random.rand(n_points)
    area = (30 * np.random.rand(n_points))**2

    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, s=area, c=colors, alpha=0.5)
    plt.title('Interactive Scatter Plot')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

interact(plot_scatter, n_points=(10, 100, 10))
```

```python
# Example 3: Bar chart with adjustable bar heights
def plot_bar_chart(bar_height):
    x = np.arange(5)
    y = np.random.rand(5) * bar_height

    plt.figure(figsize=(8, 6))
    plt.bar(x, y)
    plt.xticks(x, ['A', 'B', 'C', 'D', 'E'])
    plt.title('Interactive Bar Chart')
    plt.xlabel('Categories')
    plt.ylabel('Values')
    plt.grid(True)
    plt.show()

interact(plot_bar_chart, bar_height=(1, 10, 0.5))
```

```python
# Example 4: Line plot with adjustable line style and color
def plot_line_plot(line_color, line_style):
    x = np.linspace(0, 10, 100)
    y = np.sin(x)

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, linestyle=line_style, color=line_color)
    plt.title('Interactive Line Plot')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

interact(plot_line_plot, line_color=['blue', 'green', 'red'], line_style=['-', '--', '-.', ':'])
```

```python
# Example 5: Pie chart with adjustable slice sizes
def plot_pie_chart(slice1, slice2, slice3):
    labels = ['Slice 1', 'Slice 2', 'Slice 3']
    sizes = [slice1, slice2, slice3]

    plt.figure(figsize=(8, 6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title('Interactive Pie Chart')
    plt.axis('equal')
    plt.show()

interact(plot_pie_chart, slice1=(1, 50, 5), slice2=(1, 50, 5), slice3=(1, 50, 5))
```

```python
# Example 6: Histogram with adjustable bins
def plot_histogram(num_bins):
    data = np.random.randn(1000)

    plt.figure(figsize=(8, 6))
    plt.hist(data, bins=num_bins)
    plt.title('Interactive Histogram')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

interact(plot_histogram, num_bins=(10, 50, 5))
```

```python
# Example 7: Contour plot with adjustable levels
def plot_contour_plot(level):
    x = np.linspace(-3.0, 3.0, 100)
    y = np.linspace(-3.0, 3.0, 100)
    X, Y = np.meshgrid(x, y)
    Z = np.sin(np.sqrt(X**2 + Y**2))

    plt.figure(figsize=(8, 6))
    plt.contour(X, Y, Z, levels=np.linspace(-1, 1, level))
    plt.colorbar()
    plt.title('Interactive Contour Plot')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

interact(plot_contour_plot, level=(5, 20, 1))
```

```python
# Example 8: Quiver plot with adjustable arrow density
def plot_quiver_plot(density):
    X, Y = np.meshgrid(np.arange(-2, 2, 0.2), np.arange(-2, 2, 0.2))
    U = -1 - X**2 + Y
    V = 1 + X - Y**2

    plt.figure(figsize=(8, 6))
    plt.quiver(X, Y, U, V)
    plt.title('Interactive Quiver Plot')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

interact(plot_quiver_plot, density=(5, 20, 1))
```
